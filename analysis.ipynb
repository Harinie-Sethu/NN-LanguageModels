```python
import matplotlib.pyplot as plt

epochs = range(1, 11)  # Number of epochs
model1_train_loss = [5.4795, 4.9634, 4.8090, 4.6961, 4.6036, 4.5087, 4.4196, 4.3354, 4.2562, 4.1835]
model1_val_loss = [5.2432, 5.2085, 5.2162, 5.2277, 5.2553, 5.2635, 5.2864, 5.3135, 5.3386, 5.3654]

model2_train_loss = [5.7433, 4.9438, 4.6128, 4.3506, 4.1098, 3.8805, 3.6602, 3.4569, 3.2669, 3.0906]
model2_val_loss = [5.2504, 5.0098, 4.9895, 5.0253, 5.0997, 5.2106, 5.3541, 5.4913, 5.5996, 5.7658]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs, model1_train_loss, label='Model 1 Train', marker='o')
plt.plot(epochs, model1_val_loss, label='Model 1 Validation', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Model 1 Training and Validation Loss Over Epochs')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs, model2_train_loss, label='Model 2 Train', marker='o')
plt.plot(epochs, model2_val_loss, label='Model 2 Validation', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Model 2 Training and Validation Loss Over Epochs')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
